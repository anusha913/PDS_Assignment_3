```python
#**TASK A**

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\User\Desktop\FALL_SEM\PDS\Assignment_3\Data_clean\cleaned_diabetes.csv")

# Set a seed for reproducibility
np.random.seed(42)

# Took a random sample of 25 observations
sample_size = 25
sample = df.sample(sample_size)

# Calculate mean and highest Glucose values for the sample
sample_mean_glucose = sample['Glucose'].mean()
sample_highest_glucose = sample['Glucose'].max()

# Calculated population statistics for Glucose
population_mean_glucose = df['Glucose'].mean()
population_highest_glucose = df['Glucose'].max()

# Create a bar chart for comparison
labels = ['Population Mean', 'Population Highest', 'Sample Mean', 'Sample Highest']
values = [population_mean_glucose, population_highest_glucose, sample_mean_glucose, sample_highest_glucose]

plt.bar(labels, values, color=['blue', 'blue', 'orange', 'orange'])
plt.ylabel('Glucose Values')
plt.title('Comparison of Glucose Statistics: Population vs. Sample')
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
#**TASK B**

# Calculate the 98th percentile of BMI for the sample
sample_98th_percentile_bmi = np.percentile(sample['BMI'], 98)

# Calculate the 98th percentile of BMI for the population
population_98th_percentile_bmi = np.percentile(df['BMI'], 98)

# Create a bar chart for comparison
labels = ['Population 98th Percentile BMI', 'Sample 98th Percentile BMI']
values = [population_98th_percentile_bmi, sample_98th_percentile_bmi]

plt.bar(labels, values, color=['blue', 'orange'])
plt.ylabel('BMI Values')
plt.title('Comparison of 98th Percentile BMI: Population vs. Sample')
plt.show()
```

```python
#** TASK C**

# Number of bootstrap samples and sample size
num_of_samples = 500
each_sample_size = 150

# Initialize array for bootstrap statistics
bootstrap = np.zeros((num_of_samples, 3))

# Perform bootstrap sampling
for i in range(num_of_samples):
    sample_data = df['BloodPressure'].sample(n=each_sample_size, replace=True)
    bootstrap[i, 0] = sample_data.mean()
    bootstrap[i, 1] = sample_data.std()
    bootstrap[i, 2] = np.percentile(sample_data, 90)

# Calculate population statistics for BloodPressure
total_data_bp_mean = df['BloodPressure'].mean()
total_data_bp_std = df['BloodPressure'].std()
total_data_bp_percentile = np.percentile(df['BloodPressure'], 90)

# Create charts for comparison with different colors
plt.bar(['Bootstrap Mean', 'Total Records Mean'], [bootstrap[:, 0].mean(), total_data_bp_mean], color=['orange', 'blue'])
plt.title('BloodPressure Means')
plt.ylabel('BloodPressure')
plt.show()

plt.bar(['Bootstrap Standard Deviation', 'Total Records Standard Deviation'], 
        [bootstrap[:, 1].mean(), total_data_bp_std], color=['blue', 'orange'])
plt.title('BloodPressure Standard Deviations')
plt.ylabel('BloodPressure')
plt.show()

plt.bar(['Bootstrap Data 90th Percentile', 'Total Records 90th Percentile'], 
        [bootstrap[:, 2].mean(), total_data_bp_percentile], color=['orange', 'blue'])
plt.title('BloodPressure Data 90th Percentiles')
plt.ylabel('BloodPressure')
plt.show()
```
